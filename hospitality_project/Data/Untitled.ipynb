```python
import pandas as pd
import matplotlib.pyplot as plt

# Create clean, readable lead time bins
bins = [0, 150, 300, 450, 600, 750]
labels = ["0–150 days", "150–300 days", "300–450 days", "450–600 days", "600–750 days"]

df['lead_time_bin'] = pd.cut(df['lead_time'], bins=bins, labels=labels, include_lowest=True)

# Calculate cancellation rate for each lead-time bin
lead_cancel = df.groupby('lead_time_bin')['is_canceled'].mean() * 100

# Plot the results
plt.figure(figsize=(10,6))
lead_cancel.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Cancellation Rate by Lead Time Range', fontsize=16)
plt.xlabel('Lead Time Range', fontsize=14)
plt.ylabel('Cancellation Rate (%)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.savefig("C:/Users/vamsi/Downloads/hospitality_project/visuals/lead_time_cancellation.png", 
            dpi=300, bbox_inches='tight')


plt.show()
```

```python
import pycountry

# function to convert ISO country codes to full names
def get_country_name(code):
    if pd.isna(code) or code == "Unknown":
        return "Unknown"
    try:
        return pycountry.countries.get(alpha_3=code).name
    except:
        return code   # fallback if code not found
```

```python
top_countries = df['country'].value_counts().head(10)

# Convert codes to full names
top_countries_full = top_countries.rename(index=lambda x: get_country_name(x))
top_countries_full
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=top_countries_full.values, 
            y=top_countries_full.index, 
            palette='coolwarm')

plt.title("Top 10 Guest Countries", fontsize=16)
plt.xlabel("Number of Guests", fontsize=14)
plt.ylabel("Country", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.savefig("C:/Users/vamsi/Downloads/hospitality_project/visuals/top_countries.png", 
            dpi=300, bbox_inches='tight')

plt.show()
```
